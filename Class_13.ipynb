```python
# sinle line read
try:
    with open("file1.txt") as file1:
        filedata = file1.read()
except:
    filedata="file not fount"
print(filedata)    
```

```python
# multiple  line read from file
try:
    with open("file1.txt") as file1:
        filedata = file1.readlines()
        ff=file1.readable()
        
        print(filedata)
except:
    filedata="file not fount"
print(filedata)
```

```python
# file ko again read krna ho 1 bar ready ho jae to ni hoti read is k lea ham seek use kray gay
# seek lagae gay 
try:
    with open("file1.txt") as file1:
        filedata = file1.readlines()            
        file1.seek(0)
        data =file1.readlines() 
except:
    filedata="file not fount"
print(filedata)
print(data)
```

```python
# FOR WRITING THE FILES
# ois say purani file ka data ur jae ga 
with open('file1.txt','w') as writefile:
    writefile.writelines("hi hellow")
    writefile.writelines("hi hellow")
    
```

```python
# w+ is say read write hoga 1 sath 
# r+ k lea file ka hona lazmi hay
with open('file1.txt','w+') as writefile:
     writefile.writelines("kia hal hay")
#     writefile.read()<- is say ya read ni kar pae ga bcs is ka corsor necha ajae fa seek krna hoga hamay
    
```

```python
# file may purana data uray bagar file may new data dalnay k lea a use krna gay say data sppend ho jae ga over ride ni hoga
with open('file1.txt','a') as writefile: 
    writefile.write("hi hellow") # ab is nay purani ifle may baki data add kr dea hay
    
```
