```python
students = {'name':["Ali","Faraz","Kamran","Faizan","Nasir","Asad","Danish"],
            'Age':[20,21,24,21,20,23,np.nan],
            'Python':[89,49,78,96,100,78, 80],
            'Numpy':[78,87,np.nan,89,75,95,85],
            'Pandas':[40,75,85,48,60,89, 96]}

data = pd.DataFrame(students)
data
```

```python
#.fillna say sab change ho jae ga na wala
data.fillna(0)
data
```

```python
# is say ya hoga k ab age walay may 0 ni jae ga baki may na ki jaga 0 or inplace say ham main data may b change kr day gay
data.fillna({'Age':round(data['Age'].mean()),'Python':0,'Numpy':0,"Pandas":0},inplace=True)
data
```

```python
data['Totalobt']=data.Python+data.Pandas+data.Numpy
data
```

```python
data['Result']=['Pass' if num>200 else "Fail" for num in data.Totalobt]
data
```

```python
data['Percentage']=  round((data['Totalobt']/300)*100,2)
data
```

```python
data["Grade"]=["A+" if per >=80
              else "A" if per>=70
               else "B" if per >=60
               else "C" if per>=50
               else 'Fail'
               for per in data.Percentage
              ]
data
```

```python
data["Rechcking"]=(data.Python>=50)&(data.Pandas>=50)&(data.Numpy>=50)
data
```
