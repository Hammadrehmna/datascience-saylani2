```python
import pandas as pd
import numpy as np
frame = np.abs(
               pd.DataFrame(
                      np.random.randn(4, 3),
                      columns=list('bde'),
                      index=['Utah', 'Ohio', 'Texas', 'Oregon']))
frame
```

```python
f= lambda x:x.max() - x.min()
dt=frame.apply(f,axis='rows')
dt
```

```python
df=pd.DataFrame({
                'id':[1,2,3,4,5],
                'name':['Asad','Saad','Numi','Roman','Maria'],
                'age':[20,25,15,10,30],
                'income':[4000,7000,200,0,10000]})
df
```

```python
df['age']=df['age'].apply(lambda x: x+3)
df
```

```python
df['income']=list(map(lambda x: int(x+x*0.2),df['income'])) # 20% income barh gae is say
df
```
